```python
# Import the pandas library for data manipulation and analysi
import pandas as pd

print(pd.__version__)
```

```python
# Load the raw NYC Jobs dataset into a DataFrame
jobs = pd.read_csv("Jobs_NYC_Postings.csv")
# Display the first five rows to verify the dataset loaded correctly
jobs.head()
```

```python
# Check the number of rows and columns in the dataset
jobs.shape
```

```python
# Inspect column names, data types, and missing values
jobs.info()
```

```python
# Generate summary statistics for numerical columns
jobs.describe()
```

```python
# Count missing values in each column
jobs.isnull().sum().sort_values(ascending=False)
```

```python
# Count the number of duplicate rows
jobs.duplicated().sum()
```

```python
# Display all duplicate rows for inspection
duplicates = jobs[jobs.duplicated(keep=False)]
duplicates
```

```python
# Generate summary statistics for all categorical (text) columns
jobs.describe(include="str")
```

```python
# Verify how many rows are part of duplicate records
duplicates.shape
```

```python
# Remove exact duplicate rows to improve data quality
jobs = jobs.drop_duplicates()
```

```python
jobs.shape
```

```python
# Remove the Recruitment Contact column because it contains only missing values
jobs = jobs.drop(columns=["Recruitment Contact"])
```

```python
jobs.shape
```

```python
# Export the cleaned dataset for SQL analysis
jobs.to_csv("Jobs_NYC_Postings_Cleaned.csv", index=False)
```
